```python
#pip install chembl_webresource_client
#installed the chembl database

import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
#search for a target protein and save the results as a dataframe
target = new_client.target
target_query = target.search('coronavirus')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
#select a particular target
#want to look at the sars coronavirus 3c-like proteinase chembl3927

selected_target = targets.target_chembl_id[6]
selected_target
```

```python
#get IC50 data for the target
#get activity data for the target - seeing which compounds act upon it
activity = new_client.activity
res = activity.filter(target_chembl_id = selected_target).filter(standard_type = 'IC50')
df = pd.DataFrame.from_dict(res)
df
```

```python
#confirming that we're only dealing with ic50 data
df.standard_type.unique()
```

```python
#write to disc
#df.to_csv('data/bioactivity_data.csv', index= False)
```

```python
bioactivity_class = []
#deciding if the compound is active
for i in df['standard_value']:
    if float(i) <= 1000:
        bioactivity_class.append('active')
    elif float(i) >= 10000:
        bioactivity_class.append('inactive')
    else:
        bioactivity_class.append('intermediate')
```

```python
#making a new dataframe with only the columns of interest
selection = ['molecule_chembl_id', 'canonical_smiles', 'standard_value']
df3 = df[selection]
```

```python
df3['bioactivity_class'] = bioactivity_class
df3.head()
```

```python
# we are trying to predict drug likeness of a compound using criteria put 
# forward by Lipinski 

from rdkit import Chem
from rdkit.Chem import Descriptors, Lipinski

#function to get all the Lipinski descriptors of a drug
def lipinski(smiles, verbose=False):
    data = []
    for elem in smiles:
        mol = Chem.MolFromSmiles(elem)
        MolWt = Descriptors.MolWt(mol)
        MolLogP = Descriptors.MolLogP(mol)
        Hdonors = Lipinski.NumHDonors(mol)
        HAcceptors = Lipinski.NumHAcceptors(mol)
        data.append([MolWt, MolLogP, Hdonors, HAcceptors])

    columnnames = ['MW', 'LogP', 'NumHDonors', 'NumHAcceptors']
    descriptors = pd.DataFrame(data, columns=columnnames)
    return descriptors


df_lipinski = lipinski(df3['canonical_smiles'])
df_lipinski.head()
```

```python
df_combined = pd.concat([df3,df_lipinski], axis = 1)
```

```python
df_combined.head()
```

```python
# Convert column 'standard value' to integers
df_combined['standard_value'] = df['standard_value'].astype(float)


# this function is capping the value of the ic50 at a million for normalization purposes because 
# in the next part we're going to do a negative log transformation 

def norm_value(input):
    norm =[]
    for i in input['standard_value']:
        if i > 100000000:
            i = 100000000
        norm.append(i)
        
    input['standard_value_norm'] = norm
    
    x= input.drop('standard_value', axis =1)
    
    return x 
```

```python
import numpy as np
#this next function is converting the ic50 into negative log for ease of interpretation

def pIC50(input):
    pIC50 =[]
    for i in input['standard_value_norm']:
        molar = i *(10**-9)
        pIC50.append(-(np.log10(molar)))
    
    input['pIC50'] = pIC50
    
    x= input.drop('standard_value_norm', axis =1)
    
    return x
```

```python
df4 = norm_value(df_combined)
df4.head()
```

```python
df_final = pIC50(df4)
df_final.head()
# removing the compounds with a value of intermediate
df_analysis = df_final[df_final['bioactivity_class'] != 'intermediate']

# plotting the data to visualize the difference after negative log transformation
# this is the original data
import matplotlib.pyplot as plt
plt.hist(df3['standard_value'], bins=20, color='skyblue', edgecolor='black')
```

```python
# applying the mannwhitney test to figure out if 
# the differences between the two classes of compounds is statistically significant 

def mannwhitney(descriptor, verbose = False):
    import pandas as pd
    from numpy.random import seed
    from numpy.random import randn
    from scipy.stats import mannwhitneyu
    
    #seed the random number generator
    seed(1)
    
    selection = [descriptor, 'bioactivity_class']
    dataf = df_analysis[selection]
    active = dataf[dataf['bioactivity_class'] == 'active']
    active = active[descriptor]

    inactive = dataf[dataf['bioactivity_class'] == 'inactive']
    inactive = inactive[descriptor]
    
    #compare samples
    stat, p = mannwhitneyu(active, inactive)
    
    #interpret
    alpha = 0.05
    if p > alpha:
        interpretation = 'Non-significant'
    else:
        interpretation = 'Significant'
        
    results = pd.DataFrame({'Descriptor': descriptor,
                           'Statistics': stat, 'p': p,
                           'alpha': alpha,
                            'Interpretation':interpretation}, index =[0])
    
    filename = 'mannwhitneyu_'+descriptor+'.csv'
    results.to_csv(filename)
    return results
```

```python
mannwhitney('pIC50')
```
