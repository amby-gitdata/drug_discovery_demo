```python
# get the padel descriptors for the molecules
from padelpy import padeldescriptor

fingerprint = 'Substructure'

fp_output_file = ''.join([fingerprint, '.csv'])
fp_descriptortypes = fp[fingerprint]

padeldescriptor(mol_dir = 'molecule.smi',
               d_file = fp_output_file,
               descriptortypes = fp_descriptortypes,
               detectaromaticity = True,
               standardizenitro = True,
               standardizetautomers = True,
               threads = 2,
               removesalt = True,
               log = True,
               fingerprints = True)
```

```python
import pandas as pd
descriptors = pd.read_csv('Substructure.csv')
X = descriptors.drop('Name', axis = 1)
Y = df['class']
```

```python
#columns with low variance will be eliminated
from sklearn.feature_selection import VarianceThreshold
def remove_low_variance(input_data, threshold = 0.1 ):
    selection = VarianceThreshold(threshold)
    selection.fit(input_data)
    return input_data[input_data.columns[selection.get_support(indices = True)]]


X1 = remove_low_variance(X, threshold = 0.1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X1, Y, test_size = 0.2, random_state = 42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# build a random forest and use it to predict bioactivity state of molecule
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import matthews_corrcoef

model = RandomForestClassifier(n_estimators = 500, random_state = 42)
model.fit(X_train, y_train)

y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

mcc_train= matthews_corrcoef(y_train, y_train_pred)
print(mcc_train)

mcc_test = matthews_corrcoef(y_test, y_test_pred)
print(mcc_test)
```
